```python
import requests
import pandas as pd

# Function to fetch data from Edamam Food API (CHAMPAGNE)
def fetch_data():
    url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"
    headers = {
        "X-RapidAPI-Key": "d6802e355cmsh5476e60a3aab195p1e4acbjsnd732e83e4173",
        "X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"
    }
    params = {
        "ingr": "Champagne",
    }
    response = requests.get(url, headers=headers, params=params)
    data = response.json()
    return data

# Fetch data from API and extract relevant hints
data = fetch_data()
hints_value = data.get("hints", [])
filtered_hints = [
    hint["food"]
    for hint in hints_value
    if all(prop in hint["food"] for prop in ["foodId", "label", "category", "foodContentsLabel"])
][:10]

# Create DataFrame
df_filtered_hints = pd.DataFrame(filtered_hints)

# Specify the columns to keep
columns_to_keep = ["foodId", "label", "category", "foodContentsLabel", "image"]

# Filter DataFrame columns
df_filtered_hints = df_filtered_hints[columns_to_keep]

# Replace ";" with ","
df_filtered_hints.replace(";", ",", regex=True, inplace=True)

# Save DataFrame to CSV
file_path = "filtered_hints.csv"
df_filtered_hints.to_csv(file_path, index=False)

# Display DataFrame
display(df_filtered_hints)

# Confirmation message
print("DataFrame saved to CSV successfully.")
```
